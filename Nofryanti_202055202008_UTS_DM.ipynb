```python
import pandas as pd
import pandas_profiling
import numpy as np
import matplotlib.pyplot as plt
import os
from zipfile import ZipFile 
from scipy import stats
import seaborn as sns
```

```python
data_xlsx = pd.read_excel('/content/datasetcuaca/dataset/DatasetCuaca_convert.xlsx')
data_xlsx.to_csv('Data.csv', index=False)
```

```python
data = pd.read_csv('Data.csv')

mean = data['Suhu Udara'].mean()

median = data['Suhu Udara'].median()

mode = data['Suhu Udara'].mode()

print("Nilai Mean: ", mean)
print("Nilai Median: ", median)
print("Nilai Modus: ", mode)
```

```python
data = pd.read_csv('Data.csv')

mean = data['Curah Hujan'].mean()

median = data['Curah Hujan'].median()

mode = data['Curah Hujan'].mode()

print("Nilai Mean: ", mean)
print("Nilai Median: ", median)
print("Nilai Modus: ", mode)
```

```python
data = pd.read_csv('Data.csv')
```

```python
print(data.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Suhu Udara', y='Curah Hujan', data=data)
plt.title('Boxplot Suhu Udara dan Curah Hujan')
plt.xlabel('Suhu Udara')
plt.ylabel('Curah Hujan')
plt.show()
```

```python
sns.scatterplot(x='Suhu Udara', y='Curah Hujan', data=data)
plt.title('Scatterplot Suhu Udara dan Curah Hujan')
plt.xlabel('Suhu Udara')
plt.ylabel('Curah Hujan')
plt.show()
```

```python
std_Suhu_Udara = data['Suhu Udara'].std()

plt.figure(figsize=(8, 4))
plt.hist(data['Suhu Udara'], bins=10, alpha=0.5, color='green', edgecolor='black', label='Suhu Udara')

plt.axvline(x=std_Suhu_Udara, color='grey', linestyle='--', linewidth=3, label='Standar Deviasi Suhu Udara')

plt.title('Histogram Suhu Udara')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

```python
std_Curah_Hujan = data['Curah Hujan'].std()

plt.figure(figsize=(8, 6))
plt.hist(data['Curah Hujan'], bins=10, alpha=0.5, color='#0197f6', edgecolor='black', label='Curah Hujan')

plt.axvline(x=std_Curah_Hujan, color='orange', linestyle='--', linewidth=2, label='Standar Curah Hujan')

plt.title('Histogram Curah Hujan')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```
